```python
data = pd.read_csv("/home/jovyan/melb_data.csv")
#data.head()
y = data.Price
data.columns
cols = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
x = data[cols]

train_x, valid_x, train_y, valid_y = tts(x,y, random_state = 9)
model = dtr()
model.fit(train_x,train_y)
```

```python
print("The data we will use for predictions is this one: \n")
print(valid_x.head())
print("\n")
print("The predictions are as follows:")
preds = model.predict(valid_x)
print(preds)
print("\n")
print("The mae for this model is: " , mae(valid_y, preds)) 
```

```python
# MODEL VALIDATION:
```

```python
maes = [0,0,0,0]
k = 0
for i in [5,50,500,5000]:
    model2 = dtr(max_leaf_nodes = i, random_state =9)
    model2.fit(train_x,train_y)
    predicts_y = model2.predict(valid_x)
    maes[k] = mae(valid_y, predicts_y)
    k +=1
    print("The mae for max leaf nodes ",i, "\t","is \t",  mae(valid_y, predicts_y))

print("The best result is \t", min(maes))
```
